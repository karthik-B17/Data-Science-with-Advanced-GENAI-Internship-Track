```python
df.Score=round(df.Score).astype('Int64')
```

```python
counts=df['Batch'].value_counts()
plt.pie(
    counts.values,
    labels=counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Batch Distribution")
plt.show()
```
