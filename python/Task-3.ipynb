```python
class Solution:
    def defangIPaddr(self, address: str) -> str:
        st=""
        for each in address:
            if each=='.':
                st+=('[.]')
            else:
                st+=(each)
        return st
```

```python
class Solution:
    def findNumbers(self, nums: List[int]) -> int:
        c=0
        for each in nums:
            n=len(str(each))
            if n%2==0:
                c+=1
        return c
```

```python
class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
         return sum(nums.count(x)-1 for x in nums)//2
```

```python
class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        l=[]
        for each in range(len(nums)):
            count=0
            for i in range(len(nums)):
                if nums[each]>nums[i]:
                    count+=1
            l.append(count)
        return l

```

```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
         return eval('*'.join(str(n))) - sum(map(int, str(n)))
```

```python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        ans=0
        for each in range(n):
            ans=ans^start
            start+=2
        return ans
```
