```python
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
         nums.sort()
         return (nums[-1] - 1) * (nums[-2] - 1)
```

```python
class Solution:
    def numTeams(self, a: List[int]) -> int:
        n = len(a)
        ans = 0
        for j in range(n):
            l = sum(a[i] < a[j] for i in range(j))
            g = sum(a[i] > a[j] for i in range(j))
            r = sum(a[k] > a[j] for k in range(j+1, n))
            s = sum(a[k] < a[j] for k in range(j+1, n))
            ans += l*r + g*s
        return ans
```

```python
class Solution:
    def busyStudent(self, s: List[int], e: List[int], q: int) -> int:
        return sum(s[i] <= q <= e[i] for i in range(len(s)))
```

```python
class Solution:
    def numberOfSteps(self, n: int) -> int:
        c = 0
        while n:
            n = n//2 if n%2==0 else n-1
            c += 1
        return c
```

```python
class Solution:
    def countBits(self, n):
        ans = [0] * (n + 1)

        for i in range(1, n + 1):
            ans[i] = ans[i >> 1] + (i & 1)

        return ans
```
